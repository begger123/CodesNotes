## 展示图片

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# 采用 matplotlib 展示图片
image = mpimg.imread('plane.jpg')
plt.imshow(image)
```

```python
# 采用 matplotlib 展示图片
image = mpimg.imread('plane.jpg')
plt.axis('off')
plt.imshow(image)
```

```python
# 展示 opencv 读取的图片
import cv2
image = cv2.imread("plane.jpg")
plt.imshow(image)
```

```python
# 展示 opencv 读取的图片
import cv2
image = cv2.imread("plane.jpg")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
